```python
#Visulisation of Top 5 2022 Ranking
```

```python
Top5 = {"Country":['Finland' ,'Denmark', 'Iceland', 'Switzerland', 'Netherlands'],
        "Happiness score": [7821, 7636, 7557, 7512, 7415]}
Top5_Happiness2022 = pd.DataFrame(Top5)
print(Top5_Happiness2022)
sns.barplot(x="Happiness score", y= "Country", data=Top5, palette="flare")
plt.title("2022 Top 5 Ranked Countries")
plt.show()
```

```python
#Visulisation of Bottom 5 2022 Ranking
```

```python
Bottom5 = {"Country":['Botswana' ,'Rwanda', 'Zimbabwe', 'Lebanon', 'Afghanistan'], 
           "Happiness score": [3471, 3268, 2995, 2955, 2404
]}
Bottom5_Happiness2022 = pd.DataFrame(Bottom5)
print(Bottom5_Happiness2022)
sns.barplot(x="Happiness score", y= "Country", data=Bottom5, palette="pastel")
plt.title("2022 Bottom 5 Ranked Countries")
plt.show()
```
